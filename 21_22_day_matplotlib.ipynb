```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
```

```python
print(x)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y, label='y = x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Object Oriented Style

x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(figsize = (10, 3)) # we cereated figure and axes
ax.plot(x, x, label='linear') # y = x
ax.plot(x, x**2, label='quadratic') # y = x2
ax.plot(x, x**3, label='cubic')# y = x3
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()
```

```python
# PyPlot style

x = np.linspace(0, 2, 100)
plt.figure(figsize=(10, 3)) # we cereated figure and axes
plt.plot(x, x, label='linear') # using implicitely managed figure and axes
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

```python
fig, ax = plt.subplots(figsize = (10, 3))

x = np.arange(len(data1))
y1 = np.cumsum(data1)
ax.plot(x, y1, color='blue', linewidth=5, linestyle='--')

y2 = np.cumsum(data2)
l, = ax.plot(x, y2 , color='orange', linewidth=2)
l.set_linestyle(':')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head(20)
```

```python
titanic.info()
```

```python
titanic.set_index('PassengerId').head()
```

```python
# Create the pie-chart presenting male and female portion
```

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()
(males, females)
```

```python
portions = [males, females]
```

```python
plt.pie(
    portions,
    labels = ['Males', 'Females'],
    shadow = False,
    colors = ['blue', 'red'],
    explode = (0.15, 0),
    startangle = 90,
    autopct = "%1.1f%%"
)

# plt.axis('equal')
plt.title('Gender Portion')

plt.show()
```

```python
# Create a scatterplot for the Fare payed and the Age,
# differ the plot color by Gender
```

```python
lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex' )

lm.set(title = 'Fare x Age')

axes = lm.axes
axes[0, 0].set_ylim(-5, 600)
axes[0, 0].set_xlim(-5, 85)
```

```python
fig, ax = plt.subplots(figsize = (8, 6))

male = titanic[titanic['Sex'] == 'male']
ax.scatter(male['Age'], male['Fare'], label = 'Male', alpha = 0.6)

male = titanic[titanic['Sex'] == 'female']
ax.scatter(male['Age'], male['Fare'], label = 'Female', alpha = 0.6)

ax.set_title('Fare x Age', fontsize=16)
ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Fare', fontsize=12)
ax.set_xlim(-5, 85)
ax.set_ylim(-5, None)

ax.legend(title='Sex')

plt.show()
```

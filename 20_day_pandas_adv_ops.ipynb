```python
import numpy as np
import pandas as pd
```

```python
titatic = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv')
```

```python
air_quality = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/refs/heads/main/doc/data/air_quality_long.csv', index_col="date.utc", parse_dates=True)
```

```python
titatic.info()
```

```python
# I want to sort the Titanic data according to the age of the passengers.

titatic.sort_values(by="Age").head()
```

```python
# I want to sort the Titanic data according to the cabin class and age in descending order.

titatic.sort_values(by=['Pclass', 'Age'], ascending = False).head()
```

```python
air_quality.info()
```

```python
# We focus on data and only use the first two measurements of each location (i.e. the head of each group).
# The subset of data will be called no2_subset.
```

```python
air_quality.tail()
```

```python
air_quality.head(100)
```

```python
no2 = air_quality[air_quality["parameter"] == "no2"]
```

```python
no2.head()
```

```python
no2_subset = no2.sort_index().groupby(["location"]).head(2)
```

```python
no2_subset
```

```python
# I want the values for the three stations as separate columns next to each other.

pivoted_df = no2_subset.pivot(columns="location", values="value")
```

```python
pivoted_df
```

```python
pivoted_df.info()
```

```python
air_quality.pivot_table(
    values="value",
    index="location",
    columns="parameter",
    aggfunc="sum",
    margins=True
)
```

```python
air_quality.groupby(["parameter", "location"])[["value"]].mean()
```

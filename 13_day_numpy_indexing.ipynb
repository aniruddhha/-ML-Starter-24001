```python
a = np.array([[1,2,3,4], [5,1,7,8], [9,10,11,12]])
a[a < 5]
print(a.shape)
```

```python
five_and_up = (a >= 5)
print(a[five_and_up])
```

```python
divisible_by_2 = a[a%2==0]
print(divisible_by_2)
```

```python
bool_lnd = a[(a > 2) & (a < 11)]
print(bool_lnd)
print(a)
```

```python
bool_lor = a[(a < 5) | (a > 10)]
print(bool_lor)
print(a)
```

```python
five_up = (a > 5) | (a == 5)
print(five_up)
```

```python
a1 = np.array([[ 1, 3],[ 5, 7],[3,3]])

a2 = np.array([ [4, 2], [6, 8],[9,9]])

print(a1.shape)
print(a2.shape)

r1 = np.vstack((a1, a2))
print(r1.shape)
print(r1)

print(np.concatenate((a1, a2)))
```

```python
r2 = np.hstack((a1, a2))
print(r2)
print(r2.shape)
```
